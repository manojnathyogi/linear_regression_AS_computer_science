```python
#importing the necessary library for regression 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

Linear Regression of percentage of students achieving grade A in Computer Science in AS level 

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cs_a_raw_data.csv")
df
```

```python
%matplotlib inline
plt.xlabel("grade_threshold_a (Computer Science)")
plt.ylabel("%_achieving_a (Computer Science)")
plt.scatter(df.grade_threshold_a, df.percent_achieving_a, color='red', marker='+')
```

```python
cs_a_reg = linear_model.LinearRegression()
cs_a_reg.fit(df[['grade_threshold_a']], df.percent_achieving_a)
```

```python
cs_a_reg.predict([[98]])
```

```python
cs_a_reg.coef_
```

```python
cs_a_reg.intercept_
```

```python
prediction_a = -0.35638254 * 98 + 53.48388682611245
prediction_a
```

```python
predicted_percent_acheiving_a = cs_a_reg.predict(df[['grade_threshold_a']])
predicted_percent_acheiving_a
```

```python
%matplotlib inline
plt.xlabel("grade_threshold_a (Computer Science)")
plt.ylabel("%_achieving_a (Computer Science)")
plt.scatter(df.grade_threshold_a, df.percent_achieving_a, color='red', marker='+')
plt.plot(df.grade_threshold_a, cs_a_reg.predict(df[['grade_threshold_a']]), color='blue')
```

```python
gt_of_a = df.grade_threshold_a
gt_of_a
```

```python
d =pd.read_csv("/content/drive/MyDrive/Colab Notebooks/cs_predicted_a.csv")
d
```

```python
predicted_percent_acheiving_a = cs_a_reg.predict(d)
```

```python
d['predicted_percent_acheiving_a'] = predicted_percent_acheiving_a
d
```

```python
d.to_csv('/content/drive/MyDrive/Colab Notebooks/cs_predicted_data_a.csv', index=False)
```
